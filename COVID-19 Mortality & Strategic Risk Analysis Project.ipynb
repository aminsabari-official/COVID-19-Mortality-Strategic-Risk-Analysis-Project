```python
# ==================================================================================
# PROJECT: COVID-19 GLOBAL TREND ANALYTICS & RECOVERY FORECASTING
# Created by: Amin Sabari
# ==================================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. DATA COLLECTION (Sourcing data dari Johns Hopkins University)
print("--- Mengunduh Data Real-time ---")
url = 'https://raw.githubusercontent.com/datasets/covid-19/master/data/countries-aggregated.csv'
df = pd.read_csv(url)
print("Data berhasil dimuat.\n")

# 2. DATA PROCESSING & CLEANING
# Mengubah format tanggal dan menghitung statistik per negara
df['Date'] = pd.to_datetime(df['Date'])
country_focus = 'Indonesia'
data_id = df[df['Country'] == country_focus].copy()

# Menghitung metrik penting (New Cases & Recovery Rate)
data_id['New_Cases'] = data_id['Confirmed'].diff().fillna(0)
data_id['Recovery_Rate'] = (data_id['Recovered'] / data_id['Confirmed'] * 100).fillna(0)

# 3. DATA VISUALIZATION
plt.figure(figsize=(15, 6))
sns.set_style("whitegrid")

# Visualisasi Tren Kasus Terkonfirmasi
plt.plot(data_id['Date'], data_id['Confirmed'], color='#2c3e50', linewidth=2.5, label='Total Terkonfirmasi')
plt.fill_between(data_id['Date'], data_id['Confirmed'], color='#3498db', alpha=0.2)

plt.title(f'Visualisasi Tren Akumulasi COVID-19 - {country_focus}', fontsize=16, fontweight='bold')
plt.xlabel('Periode Waktu', fontsize=12)
plt.ylabel('Jumlah Kasus', fontsize=12)
plt.legend(loc='upper left')
plt.show()

# 4. KESIMPULAN STRATEGIS (EXECUTIVE SUMMARY)
latest_data = data_id.iloc[-1]
total_confirmed = latest_data['Confirmed']
total_recovered = latest_data['Recovered']
avg_recovery = data_id['Recovery_Rate'].mean()

print("\n" + "="*80)
print(f"{'FINAL REPORT & STRATEGIC CONCLUSION':^80}")
print("="*80)
print(f"Kandidat Analyst : Amin Sabari")
print(f"Negara Fokus     : {country_focus}")
print("-" * 80)
print(f"1. Total Akumulasi Kasus     : {total_confirmed:,.0f} orang")
print(f"2. Total Kesembuhan          : {total_recovered:,.0f} orang")
print(f"3. Rata-rata Recovery Rate   : {avg_recovery:.2f}%")
print("-" * 80)
print("KESIMPULAN ANALISIS:")
print("• Data menunjukkan pola pertumbuhan yang memerlukan manajemen risiko yang ketat.")
print("• Kemampuan dalam mengolah dataset publik ini membuktikan kompetensi dalam")
print("  Data Sourcing, Data Integrity, dan Visualisasi Informasi.")
print("• Logika analisis ini siap diimplementasikan untuk memantau performa bisnis")
print("  dan efisiensi operasional di Bank Sinarmas.")
print("="*80)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. DATA COLLECTION
url = 'https://raw.githubusercontent.com/datasets/covid-19/master/data/countries-aggregated.csv'
df = pd.read_csv(url)
df['Date'] = pd.to_datetime(df['Date'])

# 2. CHECK & FIX KEYS (Memastikan kolom benar)
# Terkadang dataset menggunakan 'Death' atau 'Deaths', kita standarkan ke 'Deaths'
if 'Death' in df.columns:
    df.rename(columns={'Death': 'Deaths'}, inplace=True)

# 3. DATA PROCESSING (Fokus pada Indonesia & Korban Jiwa)
country_focus = 'Indonesia'
data_id = df[df['Country'] == country_focus].copy()

# Menghitung metrik korban jiwa secara akurat
data_id['New_Cases'] = data_id['Confirmed'].diff().fillna(0)
data_id['Daily_Deaths'] = data_id['Deaths'].diff().fillna(0)
data_id['Fatality_Rate'] = (data_id['Deaths'] / data_id['Confirmed'] * 100).fillna(0)

# 4. VISUALIZATION
plt.figure(figsize=(16, 10))
sns.set_style("whitegrid")

# Subplot 1: Tren Kumulatif Kematian
plt.subplot(2, 1, 1)
plt.plot(data_id['Date'], data_id['Deaths'], color='#c0392b', linewidth=3, label='Total Korban Jiwa (Deaths)')
plt.fill_between(data_id['Date'], data_id['Deaths'], color='#e74c3c', alpha=0.3)
plt.title(f'Analisis Kumulatif Korban Jiwa COVID-19 - {country_focus}', fontsize=16, fontweight='bold')
plt.ylabel('Jumlah Jiwa', fontsize=12)
plt.legend(loc='upper left')

# Subplot 2: Perbandingan Harian
plt.subplot(2, 1, 2)
plt.bar(data_id['Date'], data_id['New_Cases'], color='#3498db', alpha=0.4, label='Kasus Baru Harian')
plt.bar(data_id['Date'], data_id['Daily_Deaths'], color='#2c3e50', alpha=0.9, label='Kematian Baru Harian')
plt.title('Korelasi Kasus Baru vs Angka Kematian Harian', fontsize=14, fontweight='bold')
plt.xlabel('Periode Waktu', fontsize=12)
plt.ylabel('Jumlah Orang', fontsize=12)
plt.legend()

plt.tight_layout()
plt.show()

# 5. KESIMPULAN STRATEGIS
latest = data_id.iloc[-1]
print("\n" + "="*80)
print(f"{'REPORT ANALISIS KORBAN JIWA & RISIKO FATALITAS':^80}")
print(f"{'Created by: Amin Sabari':^80}")
print("="*80)
print(f"Total Korban Jiwa (Kumulatif) : {latest['Deaths']:,.0f} jiwa")
print(f"Case Fatality Rate (CFR)      : {latest['Fatality_Rate']:.2f}%")
print(f"Total Kasus Terkonfirmasi     : {latest['Confirmed']:,.0f} orang")
print("-" * 80)
print("KESIMPULAN:")
print(f"Berdasarkan data, tingkat fatalitas adalah {latest['Fatality_Rate']:.2f}%.")
print("Koreksi data dilakukan secara otomatis untuk menjaga integritas laporan.")
print("Proyek ini membuktikan kemampuan dalam Data Cleaning dan Risk Assessment.")
print("="*80)
```
